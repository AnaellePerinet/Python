# Utilisation de pandas, matplotlib et seaborn pour les cryptomonnaies

Chargons les packages dont nous allons avoir besoin :

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

Chargons un jeu de données que nous allons appeler "bitcoin" qui provient du site finance.yahoo.com : 

```python
bitcoin = pd.read_csv('BTC-EUR.csv', index_col = 'Date', parse_dates = True)
```

Affichons un aperçu de ce jeu de données, bitcoin : 

```python
bitcoin.head()
```

Créons un graphique à l'aide du package matplotlib.pyplot : 

```python
bitcoin['Close'].plot(figsize = (9, 6))
plt.show()
```

On observe le à quel montant le bitcoin en euros s'élevait à chaque date.

```python
bitcoin.index
```

Créons un graphique à l'aide du package matplotlib.pyplot :

```python
bitcoin['2022-04']['Close'].plot()
```

On Observe le montant du bitcoin en euros en fonction des dates du mois d'Avril 2022.

On peut voir qu'il a été au plus aux alentours du 05/04/2022 et au plus bas aux alentours du 26/04/2022.

Créons un graphique à l'aide du package matplotlib.pyplot :

```python
bitcoin['2020':'2022']['Close'].plot()
```

On Observe le montant du bitcoin en euros en fonction des dates entre le mois de janvier 2020 et le mois de mai 2022.

On peut voir qu'il était très bas au mois de janvier 2020 puis qu'il a connu deux pics aux alentours du mois d'avril 2021 et de décembre 2021 avant de redescndre à nouveau.

Créons un graphique à l'aide du package matplotlib.pyplot :

```python
bitcoin.loc['2021', 'Close'].resample('M').plot()
plt.show()
```

On observe le cours du bitcoin sur l'année 2021. Les différents mois de l'année sont représentés par différentes couleurs. On note que sa valeur était basse aux mois de janvier, de juin et de juillet 2021.

Créons un graphique à l'aide du package matplotlib.pyplot :

```python
bitcoin.loc['2021', 'Close'].resample('W').mean().plot()
plt.show()
```

On observe le cours moyen du bitcoin lissé par semaine du mois de janvier 2021 au mois de janvier 2022.

Créons un graphique à l'aide du package matplotlib.pyplot :

```python
bitcoin.loc['2021', 'Close'].resample('2W').std().plot()
plt.show()
```

On observe l'écart-type du cours du bitcoin lissé toutes les deux semaines entre janvier 2021 et janvier 2022.

Créons une figure à l'aide du package matplotlib.pyplot :

```python
plt.figure(figsize = (12, 8))
bitcoin.loc['2021']['Close'].plot()
bitcoin.loc['2021', 'Close'].resample('M').mean().plot(label = "Moyenne par mois", lw = 3, ls = ':', alpha = 0.8)
bitcoin.loc['2021', 'Close'].resample('W').mean().plot(label = "Moyenne par semaine", lw = 2, ls = '--', alpha = 0.8)
plt.legend()
plt.show()
```

On observe sur la même figure trois 'graphiques' différents.
La courbe bleu représente le cours du bitcoin sur l'année 2021, la courbe orange représente le cours moyen par mois du bitcoin sur l'année 2021 et enfin la courbe verte représente le cours moyen par semaine du bitcoin sur l'année 2021. 

On affiche la moyenne, l'écart-type, le minimum et le maximum du cours du bitcoin par semaine sur l'année 2021 : 

```python
bitcoin.loc['2021','Close'].resample('W').agg(['mean', 'std', 'min', 'max'])
```

On peut par exemple voir que la valeur et la moyenne maximum ont été atteintes la semaine du 14/11/2021.
La valeur minimum du bitcoin a été elle en revanche observée sur la toute première semaine de janvier 2021.

Créons une figure à l'aide du package matplotlib.pyplot :

```python
m = bitcoin['Close'].resample('W').agg(['mean', 'std', 'min', 'max'])

plt.figure(figsize = (12, 8))
m['mean']['2021'].plot(label = "Moyenne par semaine")
plt.fill_between(m.index, m['max'], m['min'], alpha = 0.2, label = "min-max par semaine")

plt.legend()
plt.show()
```

On observe la courbe de la moyenne de la valeur du bitcoin par semaine sur l'année 2021 qui se retrouve dans l'intervalle min-max par semaine de la valeur du bitcoin.

Créons une figure à l'aide du package matplotlib.pyplot :

```python
plt.figure(figsize = (12, 8))
bitcoin.loc['2021-05', 'Close'].plot()
bitcoin.loc['2021-05', 'Close'].rolling(window = 7).mean().plot(label = "Non Centré", lw = 3, ls = '--')
bitcoin.loc['2021-05', 'Close'].rolling(window = 7, center = True).mean().plot(label = "Centré", lw = 2, ls = ':')
plt.legend()
plt.show()
```

Ce graphique produit 3 courbes :
    
    - La valeur du bictoin en mai 2021 en bleu,
    - la velur en fenêtre glissante de 7 jours non centrée en mai 2021 en orange,
    - la velur en fenêtre glissante de 7 jours centrée en mai 2021 en vert.

Créons une figure à l'aide du package matplotlib.pyplot :

```python
plt.figure(figsize = (12, 8))
bitcoin.loc['2021-05', 'Close'].plot()
for i in np.arange(0.2, 1, 0.2) :
    bitcoin.loc['2021-05', 'Close'].ewm(alpha = i).mean().plot(label = f'ewm {i}', ls = '--', alpha = 0.8)
plt.legend()
plt.show()
```

On observe sur ce graphique 5 courbes distinctes :

    - la valeur du bitcoin en mai 2021 en bleu 
    - la valeur expenentiellement pondérée par 20% du bitcoin en mai 2021 en orange
    - la valeur expenentiellement pondérée par 40% du bitcoin en mai 2021 en vert 
    - la valeur expenentiellement pondérée par 60% du bitcoin en mai 2021 en rouge 
    - la valeur expenentiellement pondérée par 80% du bitcoin en mai 2021 en violet 

Chargons un jeu de données que nous allons appeler "ethereum" qui provient du site finance.yahoo.com :

```python
ethereum = pd.read_csv('ETH-EUR.csv', index_col = 'Date', parse_dates = True)
```

Créons un graphique à l'aide du package matplotlib.pyplot :

```python
ethereum['2021']['Close'].plot()
```

On observe le cours de l'éthereum sur l'année 2021.
On voit qu'il connaît un pic entre les mois de mai et de juin avant de chuter puis que remonter à partir du mois de novembre 2021.

Créons une table qui est la fusion de notre table "bitcoin" avec notre table "ethereum". Cette fusion s'effectue sur la variable Date.

```python
btc_eth = pd.merge(bitcoin, ethereum, on = 'Date', how = 'inner', suffixes = ('_btc', '_eth'))
btc_eth
```

Créons un graphique composé de deux sous-graphiques à l'aide du package matplotlib.pyplot :

```python
btc_eth[['Close_btc', 'Close_eth']].plot(subplots = True, figsize = (12, 8))
```

On observe les cours du bitcoin et de l'ethereum entre fin 2017 et aujourd'hui (mai 2022). On peut observer que ces deux cryptomonnaies suivent à peut près les mêmes tendances mais pas à la même échelle. La valeur de l'ethereum reste beaucoup plus faible.

On établit la corrélation entre l'évolution du bitcoin et celle de l'éthereum : 

```python
correlations = btc_eth[['Close_btc', 'Close_eth']].corr()
print(correlations)
```

Comme nous avions pu le remarquer à la suite du graphique précédent, ces deux cryptomonnaies sont fortement corrélés, à hauteur de 93,1%.

Créons une 'carte des corrélations' à l'aide du package seaborn :

```python
sns.heatmap(correlations)
```

Cette figure confirme ce que nous avons dit précédemment concernant la corrélation, cependant au vu du très petit nombre de variables, cette figure ne nous apprend pas beacoup de choses supplémentaires.
